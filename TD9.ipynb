```python
def Mystere(n):
    F=factor(n)
    if len(F)<=1 :
        return false
    for f in F :
        if f[1]>1 :
            return false
    for f in F :
        if (n-1) % (f[0]-1) != 0 :
            return false
    return true
```

```python
def Carmichael(N):
    L=[]
    i=1
    while(len(L)<N):
        if(Mystere(i)):
            L.append(i)
        i=i+1
    return L
```

```python
def Fermat(n,a):
    N=IntegerModRing(n)
    if(N(a)^(n-1))!=1:
        return True #composé
    return False 
```
